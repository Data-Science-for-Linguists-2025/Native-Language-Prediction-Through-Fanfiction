```python
import spacy
import pickle
import nltk
```

```python
fics_df = pickle.load(open('fics_df.pkl', 'rb'))
nlp = spacy.load("en_core_web_sm")
```

*how to make bigrams... because you have the memory of a pickle... and not the file kind*  
`toks = [x.lower() for x in nltk.word_tokenize(y)]`  
`tokfd = nltk.FreqDist(toks)`  
`bigrams = list(nltk.bigrams(toks))`  
`bifd =nltk.FreqDist(bigrams)`  
`cfd = nltk.ConditionalFreqDist(bigrams)`  

```python
fics_df['POS'] = fics_df.work.map(lambda x: [y.pos_ for y in nlp(x)])
```

```python
fics_df['work_POS'] = fics_df.work.map(lambda x: [(y, y.pos_) for y in nlp(x)])
```
